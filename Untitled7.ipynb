```python
import sqlite3 #Passo 1: Importar o SQLite
import pandas as pd #importando o pandas
import matplotlib.pyplot as plt
import seaborn as sns

#Passo 1.1: Conectar ao banco de dados SQLite (ou criá-lo se não existir)

conexao = sqlite3.connect("dados_vendas.db")

#Passo 1.2: Criar um cursor
cursor = conexao.cursor()

#Passo 1.3: Criar uma tabela(se não existir)
cursor.execute('''
CREATE TABLE IF NOT EXISTS vendas2 (
id_vendas INTEGER PRIMARY KEY AUTOINCREMENT,
data_venda DATE,
produto TEXT,
categoria TEXT,
valor_venda REAL
)
''')

```

```python
#Passo 1.4: Inserir alguns dados
cursor.execute('''
INSERT INTO vendas2 (data_venda, produto, categoria, valor_venda) VALUES
('2023-01-01', 'Produto A', 'Eletrônicos', 1500.00),
('2023-01-05', 'Produto B', 'Roupas', 350.00),
('2023-02-10', 'Produto C', 'Eletrônicos', 1200.00),
('2023-03-15', 'Produto D', 'Livros', 200.00),
('2023-03-20', 'Produto E', 'Eletrônicos', 800.00),
('2023-04-02', 'Produto F', 'Roupas', 400.00),
('2023-05-05', 'Produto G', 'Livros', 150.00),
('2023-06-10', 'Produto H', 'Eletrônicos', 1000.00),
('2023-07-20', 'Produto I', 'Roupas', 600.00),
('2023-08-25', 'Produto J', 'Eletrônicos', 700.00),
('2023-09-30', 'Produto K', 'Livros', 300.00),
('2023-10-05', 'Produto L', 'Roupas', 450.00),
('2023-11-15', 'Produto M', 'Eletrônicos', 900.00),
('2023-12-20', 'Produto N', 'Livros', 250.00);
''')

#Passo 1.5: Confirmar as mudanças
conexao.commit()


```

```python

#Criando um DataFrame com a tabela utilizando o query para automatizar o proceso
query = "SELECT * FROM vendas2"
df_vendas = pd.read_sql_query(query, conexao)

print(df_vendas)

```

```python
sns.set(style="whitegrid", font_scale=0.6)

fig, ax = plt.subplots(1, 3, figsize=(30, 5))

bar1 = sns.barplot(data=df_vendas, x='produto', y='valor_venda', ax=ax[0], width=0.5, estimator=sum)
bar1.bar_label(bar1.containers[0], fontsize=10);
#Comparação de venda total por produto

bar2 = sns.barplot(data=df_vendas, x='data_venda', y='valor_venda', ax=ax[1], width=0.5, estimator=sum)
bar2.bar_label(bar2.containers[0], fontsize=10);
#Comparação de venda total por data


bar3 = sns.barplot(data=df_vendas, x='categoria', y='valor_venda', ax=ax[2], width=0.5, estimator=sum, errorbar=None)
bar3.bar_label(bar3.containers[0], fontsize=10);
#Valor de venda total por categoria


```
